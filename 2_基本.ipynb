```python
import akshare as ak

#debug
'''
import pandas as pd
data = {
    'code': ['600737'],
    'name': ['中粮糖业']
}
df = pd.DataFrame(data)
filename = 'temp.csv'
'''
df['行业'] = ''
df['上市时间'] = ''
for idx, row in df.iterrows():
    print('.', end = '')
    if (idx + 1) % 80 == 0: print('')
    
    code = row['code']
    try:
        df_info = ak.stock_individual_info_em(symbol = code)
        #print(df_info)
        df.at[idx, '行业'] = df_info.loc[df_info['item'] == '行业', 'value'].values[-1]  
        df.at[idx, '上市时间'] = df_info.loc[df_info['item'] == '上市时间', 'value'].values[-1]
    except Exception as e:
        print('')
        print(idx, code, e)
        df.drop(idx, inplace = True)

print('\r\nDone', len(df))
df.to_csv(filename, index = False)
df.head(3)
```
